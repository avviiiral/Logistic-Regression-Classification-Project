```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv(r"D:\ML Project1\Logistic Regression Classification Project\data.csv")
```

```python
data.drop(columns=["Unnamed: 32","id"],inplace=True)
```

```python
data.head()
```

```python
data.diagnosis = [1 if value == "M" else 0 for value in data.diagnosis]
```

```python
#divite into target variableand predictors
y= data["diagnosis"]   #out target
X = data.drop(["diagnosis"], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
#create scaler object
scaler = StandardScaler()

#fit scaler into data and tranform the data
X_scaled = scaler.fit_transform(X)

```

```python
X_scaled
```

```python
##split data
```

```python
from sklearn.model_selection import train_test_split
X_train ,X_test, y_train , y_test = train_test_split(X_scaled, y,test_size=0.3, random_state=42)
```

```python
## train model 
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
#create lr model
lr = LogisticRegression()
```

```python
## train model on data
lr.fit(X_train , y_train)

## predict target variable on test data 
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
# evaluting data
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)
print(f"accuracy:{accuracy:.2f}")
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
